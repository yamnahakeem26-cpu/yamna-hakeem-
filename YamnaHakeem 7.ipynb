Q2. Create a Python function called currency_converter() that takes two arguments:

- amount: The amount of money to be converted (in USD).
- exchange_rate (default is 280): The conversion rate from USD to another currency (e.g., PKR). You can assume the default rate is 1 USD = 280 PKR.
- 
The function should return a message showing how much the given amount is worth in the target currency (PKR).

Expected Output:

'100 USD is equivalent to 28000 PKR'

```python
def currency_converter(amount_in_USD,exchange_rate=280):
    amount_in_PKR=amount_in_USD*exchange_rate
    return (f'{amount_in_USD} USD is equivalent to {amount_in_PKR} PKR') 

currency_converter(100)    
```

Q3. Write a Python function display_student_info that accepts the following details about a student:

- name (Keyword argument): The name of the student.
- age (Keyword argument): The age of the student.
- grade (Keyword argument): The grade the student is in.
- school (Keyword argument): The name of the school the student attends.

Expected Output:

- Name: John Doe
- Age: 15
- Grade: 10th
- School: Greenwood High

```python
def display_student_info(**kargs):
    for x,y in kargs.items():
        print(x.capitalize(),':',y)

display_student_info(name='John Doe',age=16,grade='10th',school='Greenwood High')
```

Q5. Write a Python function calculate_rectangle that accepts the length and width of a rectangle as arguments. Inside this function, create two nested functions:

- calculate_area: This function should calculate the area of the rectangle.
- calculate_perimeter: This function should calculate the perimeter of the rectangle.
- The main function should return both the area and the perimeter of the rectangle.

Expected Output:

Area: 15

Perimeter: 16


```python
def calculate_rectangle(length,width):
    def calculate_area():
        return length*width
    
    def calculate_perimeter():
        return 2*(length+width)

    return f'Area: {calculate_area()},\nperimeter: {calculate_perimeter()}'

print(calculate_rectangle(3,5))    
```

Write a Python program that:

Has a balance of 1000.

Asks the user for a withdrawal amount.

Checks for these errors:

Non-numeric input → print "Please enter a valid number."

Amount ≤ 0 → print "Amount must be more than zero."

Amount > balance → print "Insufficient balance."

If valid, deducts amount and shows remaining balance.

Always prints "Thank you for using the ATM."

Expected Output:

Q9:

Define a class named Light with two methods:

turn_on() — prints: "The light is now ON."

check_status() — prints: "The light is bright and working."

Create an object of the Light class and call both methods.

Expected Output:

The light is now ON.

The light is bright and working.

```python
class Light():
    def turn_on(self):
        print(f'The Light is now ON')
    def check_status(self):
        print(f'The Light is bright and working')

light=Light()
light.turn_on()
light.check_status()
```

Q10:

Define a class named Student that has the following:

Attributes: name, grade, and age.

A method introduce() that prints:

"My name is {name}, I am {age} years old."

A method show_grade() that prints:

"My grade is {grade}."

Create an object of the Student class with sample data (e.g., "Alice", "A", 16), and call both methods on the object.

Expected Output:

My name is Alice, I am 16 years old.

My grade is A.

```python
class Student():
    def introduce(self,name,age):
        print(f'My name is {name}, I am {age} years old')
    def show_grade(self,grade):
        print(f'My grade is {grade}')

talibat=Student()
talibat.introduce("Yamna",26)
talibat.show_grade("A+")
```
